```python
import numpy as np 
import pandas as pd

df = pd.read_csv('../data/dengue_19_20_21.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['District','Date'])

for col in ['Temp_Mean','Rain_Sum','Humidity_Proxy']:
    for lag in[1,2,3,4]:
        df[f'{col}_lag{lag}'] = df.groupby('District')[col].shift(lag)

df['Rain_roll3'] = df.groupby('District')['Rain_Sum'].transform(lambda x: x.rolling(3).mean())
df['Temp_roll3'] = df.groupby('District')['Temp_Mean'].transform(lambda x: x.rolling(3).mean())
```

```python
df['Month'] = df['Date'].dt.month
df['Month_sin'] = np.sin(2 * np.pi * df['Month'] /12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] /12)
```

```python
df['Cases_log'] = np.log1p(df['Dengue_Cases'])
```

```python
threshold = df['Dengue_Cases'].quantile(0.75)
df['Risk_Label'] = (df['Dengue_Cases'] > threshold).astype(int)
```

```python
df.dropna(inplace=True)
df.to_csv('../data/dengue_features.csv', index=False)
```
