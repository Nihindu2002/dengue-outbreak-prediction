```python
import folium,pandas as pd,json

df = pd.read_csv('../data/dengue_features.csv')
geo = json.load(open('../data/cities-by-district.json'))

district_2021 = df[df['Date']>='2021-01-01'].groupby('District')['Dengue_Cases'].sum().reset_index()

m = folium.Map(location=[7.8731, 80.7718], zoom_start=7)
folium.Choropleth(
    geo_data = geo,
    data = district_2021,
    columns=['District', 'Dengue_Cases'],
    key_on='feature.properties.Name_2',
    fill_color='YlOrRd',
    legend_name='Dengue Cases in 2021'
).add_to(m)
m.save('../outputs/dengue_map.html')
m
```
